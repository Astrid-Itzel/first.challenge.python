# 2. Ventas por categoría

```python
import pandas as pd

# URLs de las tiendas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
]

# Procesar cada tienda
for i, url in enumerate(urls, start=1):
    df = pd.read_csv(url)
    df.columns = df.columns.str.strip()  # limpiar espacios en nombres de columnas

    ventas_por_categoria = (
        df.groupby("Categoría del Producto")["Producto"]
        .count()
        .sort_values(ascending=False)
    )

    print(f"\n Tienda {i} - Categorías más populares:")
    print(ventas_por_categoria)
```

# 4. Productos más y menos vendidos

```python

import pandas as pd
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Concatenar todos los datos
df_todas = pd.concat([tienda, tienda2, tienda3, tienda4])

# Contar ventas por tienda y producto
ventas_por_tienda_producto = df_todas.groupby(['Tienda', 'Producto']).size().reset_index(name='Ventas')

# Mostrar resultados
for tienda_nombre in ventas_por_tienda_producto['Tienda'].unique():
    print(f"\nProductos más vendidos en {tienda_nombre}:")
    df_t = ventas_por_tienda_producto[ventas_por_tienda_producto['Tienda'] == tienda_nombre]
    print(df_t.sort_values('Ventas', ascending=False).head(5))

    print(f"\nProductos menos vendidos en {tienda_nombre}:")
    print(df_t.sort_values('Ventas', ascending=True).head(5))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for tienda_nombre in ventas_por_tienda_producto['Tienda'].unique():
    df_t = ventas_por_tienda_producto[ventas_por_tienda_producto['Tienda'] == tienda_nombre]
    top5 = df_t.sort_values('Ventas', ascending=False).head(5)

    plt.figure(figsize=(8,5))
    sns.barplot(data=top5, y='Producto', x='Ventas', palette="viridis")
    plt.title(f'Top 5 productos más vendidos en {tienda_nombre}')
    plt.xlabel('Cantidad de ventas')
    plt.ylabel('Producto')
    plt.show()
```

## Introducción
El objetivo de este análisis fue determinar en cuál de las cuatro tiendas el Sr. Juan debería vender sus productos, evaluando factores clave como ingresos totales, categorías más vendidas, satisfacción del cliente, desempeño de productos y costos de envío promedio. El fin es identificar la tienda con mejor balance entre rentabilidad y competitividad.

---

## 1. Ingresos Totales por Tienda
Tras el cálculo de los ingresos totales (Precio × número de ventas), se observó que **Tienda 2** lidera en facturación, seguida muy de cerca por **Tienda 1**. Tienda 3 y Tienda 4 presentan ingresos más moderados.

> **Insight:** Tienda 2 combina alto volumen de ventas con precios competitivos, lo que explica su liderazgo en ingresos.

---

## 2. Distribución de Categorías de Productos
El análisis de categorías mostró que:
- **Electrodomésticos** y **Tecnología** son predominantes en Tienda 1 y Tienda 2.
- Tienda 3 y Tienda 4 tienen mayor proporción de productos para el hogar, muebles y artículos de cocina.
- Algunas categorías como instrumentos musicales o libros presentan menor rotación.

> 📊 **Insight:** Tienda 2 ofrece una mezcla equilibrada de categorías con alta demanda y rotación constante.

---

## 3. Calificaciones Promedio de Clientes
Las calificaciones promedio fueron:
- **Tienda 2:** 4.7 ⭐

## Conclusión y Recomendación
Tras el análisis integral de los datos, **la tienda menos eficiente es Tienda 4**, debido a que:

- Registra **los ingresos más bajos** entre todas las tiendas.
